```python
#Reto - Análisis del progreso mundial de vacunación - Learning gate
#Karla Fabiola Ramirez Martinez 

#Notas: en varios de los ejemplos puse tails o head para mostrar los resultados
#Mas que nada para que fuera mas sencillo ver la informacion
#Imports
import pandas as pd
import numpy as np
import datetime

#a. Extraer la información del archivo. 
data=pd.read_csv("drive/MyDrive/country_vaccinations.csv")

#b. Mostrar la estructura y tipos de datos de cada columna para identificar
# qué operaciones puedes realizar con cada una de ellas, asegurándote que las 
#columnas con fechas sean del tipo datetime64. 
print("*********************************************************************")
print("Tipos de datos")
df = pd.DataFrame(data)
df["date"] =  pd.to_datetime(df["date"], format="%Y/%m/%d")
result = df.dtypes
print(result)

#c. Determinar la cantidad de vacunas aplicadas de cada compañía
# (con base en cómo lo reporta cada país en la columna vaccines). 

print("*********************************************************************")
print("Cantidad de vacunas aplicadas de cada compañia")
duplicados = df.pivot_table(columns=['vaccines'], aggfunc='size')
print (duplicados)
 
#d. Obtener la cantidad de vacunas aplicadas en todo el mundo
print("*********************************************************************")
print("Cantidad de vacunas aplicadas en todo el mundo")
vacunastotal = pd.DataFrame({'total_vaccinations':[7500000.0,np.nan]})
print(vacunastotal['total_vaccinations'].sum())

#e. Calcular el promedio de vacunas aplicadas por país. 
print("*********************************************************************")
print("Promedio de vacunas aplicadas por pais")
df["avg_country"]=df.groupby('country')['total_vaccinations'].transform('mean')
vacunasavg_pais = df.pivot_table(columns=['country'])
print (vacunasavg_pais)

#f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.
print("*********************************************************************")
print("Vacunas aplicadas el día 29/01/21")
print (df.loc[df['date'] == '29/01/21', 'total_vaccinations'].sum())

#g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales 
#y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas
#daily_vaccionations y daily_vaccionations_raw.  
print("*********************************************************************EXCEL")
print("Dataframe")
df3=df
conDiferencias=df3
conDiferencias['diferencias'] = df['daily_vaccinations_raw'] - df['daily_vaccinations']
print(conDiferencias.tail())

#h. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua. 
print("*********************************************************************")
print("Periodo de tiempo de fecha mas reciente y antiguo")
maximo= df['date'].max()
minimo= df['date'].min()
print("La diferencia son : ",((maximo - minimo)/np.timedelta64(1, 'D')), " dias")

#i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada
# (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter /). 
print("*********************************************************************EXCEL")
print("conCantidad")
df2=df
df2['canVac'] =df2.groupby('date')['vaccines'].sum()
conCantidad=df2
print(conCantidad.tail())

#j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020
print("*********************************************************************EXCEL")
print("Antes20")
antes20=df.loc[df['date'] < '20/12/20']
print (antes20.head())
#k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer. 
print("*********************************************************************EXCEL")
print("Pfizer")
pfizer= df[df['vaccines'].str.contains('Pfizer')]
print(pfizer.tail())

#l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado 
#resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter. 
print("*********************************************************************EXCEL")
print("Exportar a Excel en diferentes hojas")
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
  conDiferencias.to_excel(writer, sheet_name='conDiferencias')
  conCantidad.to_excel(writer, sheet_name='conCantidad')
  pfizer.to_excel(writer, sheet_name='pfizer')
  antes20.to_excel(writer, sheet_name='antes20')


```
